```python
import pandas as pd
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
df1
```

```python
final=df1.drop(["Survived"],axis=1)
final
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender=LabelEncoder()
gender_n=le_gender.fit_transform(final["Sex"])
final['Gender']=gender_n
final
```

```python
final=final.drop(["Sex"],axis=1)
final
```

```python
med = df['Age'].median()
final['Age']=df['Age'].fillna(med)
```

```python
target=df1['Survived']
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(final,target,train_size=0.8)
```

```python
x_test
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test,y_test)
```

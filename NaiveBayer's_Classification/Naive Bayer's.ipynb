```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],inplace=True,axis=1)
df
```

```python
target=df["Survived"]
inputs=df.drop(["Survived"],axis=1)
inputs
```

```python
dumm=pd.get_dummies(df.Sex)
dumm.head()
```

```python
inputs=pd.concat([inputs,dumm],axis=1)
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
m=df.Age.mean()
inputs.Age.fillna(m,inplace=True)
```

```python
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.drop(["Sex"],axis=1,inplace=True)
```

```python
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,train_size=0.8)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

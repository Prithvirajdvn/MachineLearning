```python
from PIL import Image
import numpy as np'
import pandas as pd
```

```python
image=Image.open("images.jpeg")
img_arr=np.array(image)
print(img_arr)
```

```python
img_arr.shape
```

```python
img1d=img_arr.reshape(-1)
img1d.shape
```

```python
df=pd.DataFrame(img1d,columns=['Pixel'])
df
```

```python
import numpy as np
import pandas as pd

# Assuming 'image_array' is your NumPy array representing the image

# Reshape the image array to have channels as columns
height, width, channels = img_arr.shape
image_reshaped = img_arr.reshape(height * width, channels)

# Create column names for the DataFrame
column_names = [f"Channel_{i+1}" for i in range(channels)]

# Create a DataFrame from the reshaped array
df = pd.DataFrame(image_reshaped, columns=column_names)

# Print the DataFrame
print(df)
```

```python
import numpy as np
import pandas as pd

# Assuming 'image_array' is your NumPy array representing the image

# Flatten the image array to a 1D array
flattened_image = img_arr.flatten()

# Create a DataFrame with a single row
df = pd.DataFrame([flattened_image])

# Print the DataFrame
print(df)
```
